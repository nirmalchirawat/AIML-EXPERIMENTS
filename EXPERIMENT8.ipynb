Implement Dimensionality reduction using Principal Component
Analysis (PCA) method.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# ── Step 1: Load Dataset ───────────────────────
df = pd.read_csv("data.csv")
print("\nStep 1: Original Dataset\n")
print(df)

# ── Step 2: Select Numerical Data ──────────────
X = df[["Age", "Marks", "Height"]]
print("\nStep 2: Selected Features\n")
print(X)

# ── Step 3: Handle Missing Values ──────────────
X = X.fillna(X.mean())
print("\nStep 3: After Handling Missing Values\n")
print(X)

# ── Step 4: Standardization ────────────────────
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nStep 4: Standardized Data\n")
print(pd.DataFrame(X_scaled, columns=X.columns))

# ── Step 5: Apply PCA ──────────────────────────
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])

print("\nStep 5: PCA Result (Reduced Data)\n")
print(pca_df)

# ── Step 6: Explained Variance ─────────────────
print("\nStep 6: Explained Variance Ratio\n")
print(pca.explained_variance_ratio_)

```

```python
# ── Step 7: Final Graph ────────────────────────
plt.figure()
plt.scatter(pca_df["PC1"], pca_df["PC2"])

plt.title("PCA Visualization")
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.show()
```
