Write a Python program to demonstrate various Data Visualization
Techniques using Matplotlib and Seaborn libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ── Load Dataset ───────────────────────────────
df = pd.read_csv("visualization_data.csv")

print("\nDataset:\n", df)

# ── 1. Line Plot (Matplotlib) ──────────────────
plt.figure()
plt.plot(df["Marks"], marker='o')
plt.title("Line Plot of Marks")
plt.xlabel("Index")
plt.ylabel("Marks")
plt.show()

# ── 2. Bar Plot (Matplotlib) ───────────────────
plt.figure()
plt.bar(df["Name"], df["Marks"])
plt.title("Bar Plot of Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.show()

# ── 3. Histogram (Matplotlib) ──────────────────
plt.figure()
plt.hist(df["Marks"])
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# ── 4. Scatter Plot (Matplotlib) ───────────────
plt.figure()
plt.scatter(df["Age"], df["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()

# ── 5. Box Plot (Seaborn) ──────────────────────
plt.figure()
sns.boxplot(x=df["Marks"])
plt.title("Box Plot of Marks")
plt.show()

# ── 6. Count Plot (Seaborn) ────────────────────
plt.figure()
sns.countplot(x="Gender", data=df)
plt.title("Count Plot of Gender")
plt.show()

# ── 7. Heatmap (Seaborn) ───────────────────────
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
