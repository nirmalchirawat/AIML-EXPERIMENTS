```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

print("Original Data:\n", df)

# ── Feature Scaling ────────────────────────────
num_cols = ["Age", "Marks", "Height"]

# Min-Max Scaling
df_minmax = df.copy()
df_minmax[num_cols] = (df_minmax[num_cols] - df_minmax[num_cols].min()) / (df_minmax[num_cols].max() - df_minmax[num_cols].min())

print("\nMin-Max Scaled:\n", df_minmax)

# Z-score Scaling
df_zscore = df.copy()
df_zscore[num_cols] = (df_zscore[num_cols] - df_zscore[num_cols].mean()) / df_zscore[num_cols].std()

print("\nZ-score Scaled:\n", df_zscore)
```

```python
import matplotlib.pyplot as plt

plt.figure()

# Original Data (Marks)
plt.plot(df["Marks"], marker='o', label="Original Marks")

# Min-Max Scaled
plt.plot(df_minmax["Marks"], marker='o', label="Min-Max Marks")

# Z-score Scaled
plt.plot(df_zscore["Marks"], marker='o', label="Z-score Marks")

plt.title("Feature Scaling Comparison (Marks)")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()

plt.show()
```
