Using Python language do the exploratory data analysis of dataset
imported in the lab 4.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data2.csv")
```

```python
print("\n", df.head())
print("\n", df.tail())
print("\n", df.info())
print("\n", df.describe())
```

```python
print("Rows, Columns:", df.shape)
print("Size:", df.size)
print(df.isnull().sum())
```

```python
print("Mean:\n", df.mean(numeric_only=True))
print("Sum:\n", df.sum(numeric_only=True))
print("Min:\n", df.min(numeric_only=True))
print("Max:\n", df.max(numeric_only=True))
```

```python
df.hist()
plt.show()
```

```python
df.plot()
plt.show()
```

```python
df.boxplot()
plt.show()
```
