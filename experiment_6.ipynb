implement the missing value, and outlier handling data preprocessing
techniques on the data set imported in lab 4 or any other dataset

```python

import pandas as pd
import numpy as np

df = pd.read_csv("experiment6.csv")

numeric_df = df.select_dtypes(include=np.number)

numeric_df = numeric_df.fillna(numeric_df.mean())

df[numeric_df.columns] = numeric_df

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)

IQR = Q3 - Q1


filter_condition = ~((numeric_df < (Q1 - 1.5 * IQR)) |
                     (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)

df_cleaned = df[filter_condition]

print("Data after handling missing values and removing outliers:\n")
print(df_cleaned)
```
